```python
apps.prime_genre.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

```python
#mismo plot filtrando games
apps[apps['prime_genre'] != 'Games'].prime_genre.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

```python
#mismo plot filtrando games
apps[apps['price'] > 0].prime_genre.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

```python
#mismo plot filtrando games
apps[(apps['price'] > 0) & (apps['prime_genre'] != 'Games')].prime_genre.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

Los resultados varan proporcionalmente en el caso de las apps pagas y de las gratuitas. No perder de vista que todas las apps tambin pueden generar ingresos mediante *in app purchases*. A partir de esto se puede concluir que de querer desarrollar alguna app que genere ingresos, aparentemente lo que ms convendra hacer es desarrollar algn juego. De no ser posible por tratarse de un rea muy particular, o no ser un rea de inters, le seguiran las opciones de Entretenimiento, Educacin y Foto/video.

```python
interest_areas = ['Games', 'Education', 'Entertainment', 'Photo & Video']
```

```python
# veamos los ratings de contenido de las areas de interes
apps[apps['prime_genre'].isin(interest_areas)].cont_rating.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

```python
apps[(apps['prime_genre'].isin(interest_areas)) & (apps['cont_rating'] == '17+')].prime_genre.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

La distribucion se mantiene en el caso de las apps para pblico de 17 aos o ms.

```python
# veamos los ratings de los usuarios de las areas de interes
apps[apps['prime_genre'].isin(interest_areas)].user_rating.value_counts(ascending=True).sort_index().plot(kind='barh', figsize=(16,10))
```
